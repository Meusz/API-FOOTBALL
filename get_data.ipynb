# Funciones

Necesarias para una vez obtenido la info de la web, transformarlo en un csv manejable y leible

```python
def subdata(titulo,texto):
    #print("Se va a procesar datos de: ",titulo)
    result=[]
    datos = texto.split(",")
    for lineas in datos:
        linea = lineas.split(":")
        
        if len(linea) > 2:
            linea[1] = ":".join(linea[1:])
        result.append(linea[1])
    return result

def escribir_csv(file_name,data):
    header = ["player_id","player_name","player_photo","player_type","player_injury_reason","team_id","team_name","team_logo","fixture_id","fixture_timezone",
        "fixture_date","fixture_timestamp","league_id","league_season","league_name","league_country","league_logo","league_flag"]
    
    with open(file_name, mode='w') as employee_file:
        employee_writer = csv.writer(employee_file, delimiter=',', quotechar='"', quoting=csv.QUOTE_MINIMAL)

        employee_writer.writerow(header)
    
        jugadores = data.split("{\"player\":{")
        
        for i in range(1,len(jugadores)-1):
            result=[]
            datos = jugadores[i].split("},")
            
            result = result + subdata("player",datos[0])
            for secciones in datos[1:4]:
                seccion = secciones.split(":{")
                result = result + subdata(seccion[0],seccion[1])
            employee_writer.writerow(result)
```

```python
conn = http.client.HTTPSConnection("v3.football.api-sports.io")
season = str(2023)
league = str(39)  #140 Laliga  39 Premier League
headers = {
    'x-rapidapi-host': "v3.football.api-sports.io",
    'x-rapidapi-key': "c3bee586de515d811abbb66b94fc00cd",
    }

url = "/injuries?season="+str(season)+"&league="+str(league)

conn.request("GET", url , headers=headers)

res = conn.getresponse()
data = res.read().decode("utf-8")

file_name = "league-"+str(league)+"_season-"+str(season)+".csv"

escribir_csv(file_name,data)
print("Se ha escrito el archivo : ", file_name)



    

```
